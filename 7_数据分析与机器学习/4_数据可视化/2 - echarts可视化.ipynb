```python
from pyecharts import options as opts
from pyecharts.charts import Map
 
# 创建外层地图
def base_map(set_num, legend1, legend2, legend3, data_opts1, data_opts2, data_opts3, title):
    provinces_map = (
        Map(
            init_opts=opts.InitOpts(width ="1200px", height="600px")
        )
        .add(legend1, data_opts1, maptype="china", is_map_symbol_show=False, name_map=province_translate, zoom=1.2)
        .add(legend2, data_opts2, maptype="china", is_map_symbol_show=False, name_map=province_translate, zoom=1.2,is_selected=False)
        .add(legend3, data_opts3, maptype="china", is_map_symbol_show=False, name_map=province_translate, zoom=1.2,is_selected=False)   
        
        # 全局配置项
        .set_global_opts(
            # 配置标题
            title_opts=opts.TitleOpts(
                title=title, 
                pos_right = "center"                
            ),
            # 配置视觉条
            visualmap_opts=opts.VisualMapOpts(
                max_= max(set_num),
                min_= min(set_num),
                pos_top="bottom",     # 分段位置
                pos_left="left",
                orient="vertical",    # 分段朝向 
                range_color = ["#E0ECF8", "#045FB4"]    # 颜色填充    
               
            ),
#             # 配置提示项
#             tooltip_opts=opts.TooltipOpts(
#                 formatter = '{b}:{c}'                            
#             ),
            # 配置图例
            legend_opts=opts.LegendOpts(
                pos_left='10px',
                orient="vertical", 
                selected_mode="single"
            ),            
            
        )
    )
    return provinces_map


# 外层地图数据
province_set_num = china_province_sum.total_num
provinces_map = base_map(province_set_num, "total_num", "six_total_num", "four_total_num", china_province_total_num, china_province_six_total_num, china_province_four_total_num, "省份分布")
provinces_map.render("./map/中国地图.html")
```

```python
# 汇总中国城市数据
china_city_sum = data[data["permanent_country_cn"]=="中国"]
# 清洗城市数据
china_city_sum['city_clear'] = china_city_sum['permanent_city_cn'].map(city_translate)

# print(china_city_sum[china_city_sum['permanent_province_cn']=="西藏"])
# 获取所有省份
china_province = china_city_sum.permanent_province_cn.unique()

# 组合城市数据
china_city_total_num = {}
china_city_six_total_num = {}
china_city_four_total_num = {}
for province in china_province:
    # 获取城市分布数据
    china_city_total_num[province] = china_city_sum[china_city_sum["permanent_province_cn"]== province].loc[:, ["city_clear","total_num"]].values.tolist()
    china_city_six_total_num[province] = china_city_sum[china_city_sum["permanent_province_cn"]== province].loc[:, ["city_clear","six_total_num"]].values.tolist()
    china_city_four_total_num[province] = china_city_sum[china_city_sum["permanent_province_cn"]== province].loc[:, ["city_clear","four_total_num"]].values.tolist()  
```

```python
from pyecharts import options as opts
from pyecharts.charts import Map
 
# 创建省份地图
for province in china_province:
    # 获取分布数据
    data_opts1 = china_city_total_num[province]
    data_opts2 = china_city_six_total_num[province]
    data_opts3 = china_city_four_total_num[province]
    # 获取视觉条最大最小值
    city_set_num = pd.DataFrame(china_city_total_num[province], columns=['city','total_num']).total_num
    # 获取省份地图
    city_map = (
        Map(
            init_opts=opts.InitOpts(width ="100%", height="600px")
        )
        .add("total_num", data_opts1, maptype=province, is_map_symbol_show=False, zoom=1)
        .add("total_six_num", data_opts2, maptype=province, is_map_symbol_show=False, zoom=1,is_selected=False)
        .add("total_four_num", data_opts3, maptype=province, is_map_symbol_show=False, zoom=1,is_selected=False)   
        
        # 全局配置项
        .set_global_opts(
            # 配置标题
            title_opts=opts.TitleOpts(
                title=province + "地图", 
                pos_right = "center"                
            ),
            # 配置视觉条
            visualmap_opts=opts.VisualMapOpts(
                max_= max(city_set_num),
                min_= min(city_set_num),
                range_color = ["#E0ECF8", "#045FB4"]    # 颜色填充     
                                  
            ),
            # 配置图例
            legend_opts=opts.LegendOpts(
                pos_left='10px',
                orient="vertical"                
            ),     
        
        )
    )
    city_map.render(f"./map/{province}地图.html")
```

```python
# 修改省份地图，实现省市下钻
from bs4 import BeautifulSoup
 
def read_html(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        html_content = f.read()
    soup = BeautifulSoup(html_content, 'html.parser')
    id = soup.find("div")['id']
    return soup.prettify(), id

def write_html(out_path, text): 
    # 打开或创建一个HTML文件
    with open(out_path, 'w', encoding='utf-8') as file:
        # 写入文本
        file.write(text)
```
