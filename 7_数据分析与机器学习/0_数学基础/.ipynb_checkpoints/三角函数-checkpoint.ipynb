# 三角函数

## 一、$\pi$和弧度

$\pi$是圆的周长与直径的比值
$$
\pi = \frac{C}{R} = \frac{C}{2r}
$$

弧度是角的度量单位。它是由国际单位制导出的单位，单位缩写是rad。定义：弧长等于半径的弧，其所对的圆心角为1弧度。所以一周的弧度数为，

$$
1弧度 = \frac{r}{r}  
$$

$$
1周的弧度 = \frac{C}{r} = \frac{2\pi r}{r} = 2\pi
$$

周角为$2\pi$弧度，平角（即180°角）为$\pi$弧度，直角为$\frac{\pi}{2}$弧度。

## 二、扇形的面积

$$
S = \frac{1}{2}lr = \frac{1}{2}*\frac{2\pi r \alpha}{360}* r = \frac{\pi r^{2}}{360}*\alpha
$$

$\alpha$是扇形的角度

## 三、三角函数

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-2*np.pi,np.pi*2,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y)
plt.plot(x,z)
```
