```python
import numpy as np
import scipy as sp
from scipy.optimize import leastsq
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# ÁõÆÊ†áÂáΩÊï∞
def real_func(x):
    return np.sin(2*np.pi*x)

# Â§öÈ°πÂºè
def fit_func(p, x):
    f = np.poly1d(p)     # ÁîüÊàêÂ§öÈ°πÂºèÂáΩÊï∞Ôºånp.poly1d([1,2,3]) = 1x^2+2x^1+3x^0
    return f(x)

# ÊÆãÂ∑Æ
def residuals_func(p, x, y):
    ret = fit_func(p, x) - y
    return ret
```

```python
x = np.linspace(0,1,10)
x_points = np.linspace(0,1,1000)
_y = real_func(x)
# Â¢ûÂä†Ê≠£Â§™ÂàÜÂ∏ÉÂô™Èü≥ÁöÑÁõÆÊ†áÂáΩÊï∞ÂÄº  np.random.normal()
y = [np.random.normal(0,0.1) + y for y in _y]
# ÁªòÂõæ
plt.rcParams['font.sans-serif']=['SimHei'] #Áî®Êù•Ê≠£Â∏∏ÊòæÁ§∫‰∏≠ÊñáÊ†áÁ≠æ
plt.plot(x,_y,label='sin2œÄ')
plt.plot(x,y,label='sin2œÄ+Ê≠£Â§™Âπ≤Êâ∞')
plt.legend()
```

```python
# ÂçÅ‰∏™ÁÇπ
x = np.linspace(0, 1, 10)
x_points = np.linspace(0, 1, 1000)
# Âä†‰∏äÊ≠£ÊÄÅÂàÜÂ∏ÉÂô™Èü≥ÁöÑÁõÆÊ†áÂáΩÊï∞ÁöÑÂÄº
y_ = real_func(x)
y = [np.random.normal(0, 0.1) + y1 for y1 in y_]

def fitting(M=0):
    # ÈöèÊú∫ÂàùÂßãÂåñÂ§öÈ°πÂºèÂèÇÊï∞
    p_init = np.random.rand(M+1)
    # ÊúÄÂ∞è‰∫å‰πòÊ≥ï
    p_lsq = leastsq(residuals_func, p_init, args=(x, y))
    print('Fitting Parameters:', p_lsq[0])
    
    # ÂèØËßÜÂåñ
    plt.plot(x_points, real_func(x_points), label='real')
    plt.plot(x_points, fit_func(p_lsq[0], x_points), label='fitted curve')
    plt.plot(x, y, 'bo', label='noise')
    plt.legend()
    return p_lsq
```

## M=0

```python
p_lsq_0 = fitting(M=0)
```

## M=1

```python
p_lsq_1 = fitting(M=1)
```

## M=3

```python
p_lsq_3 = fitting(M=3)
```

## M=9

```python
p_lsq_9 = fitting(M=9)
```

## Ê≠£ÂàôÂåñ

ÁªìÊûúÊòæÁ§∫ËøáÊãüÂêàÔºå ÂºïÂÖ•Ê≠£ÂàôÂåñÈ°π(regularizer)ÔºåÈôç‰ΩéËøáÊãüÂêà

$Q(x)=\sum_{i=1}^n(h(x_i)-y_i)^2+\lambda||w||^2$„ÄÇ

ÂõûÂΩíÈóÆÈ¢ò‰∏≠ÔºåÊçüÂ§±ÂáΩÊï∞ÊòØÂπ≥ÊñπÊçüÂ§±ÔºåÊ≠£ÂàôÂåñÂèØ‰ª•ÊòØÂèÇÊï∞ÂêëÈáèÁöÑL2ËåÉÊï∞,‰πüÂèØ‰ª•ÊòØL1ËåÉÊï∞„ÄÇ

- L1: regularization\*abs(p)

- L2: 0.5 \* regularization \* np.square(p)

```python
regularization = 0.0001
def residuals_func_regularization(p, x, y):
    ret = fit_func(p, x) - y     # ËØØÂ∑Æ
    ret = np.append(ret, np.sqrt(0.5 * regularization * np.square(p)))  # L2ËåÉÊï∞
    return ret
```

```python
# ÊúÄÂ∞è‰∫å‰πòÊ≥ïÂä†Ê≠£ÂàôÂåñÈ°π
p_init = np.random.rand(9 + 1)
p_lsq_regularization = leastsq(residuals_func_regularization,p_init,args=(x,y))
```

```python
plt.plot(x_points, real_func(x_points), label='real')
plt.plot(x_points, fit_func(p_lsq_9[0], x_points), label='fitted curve')
plt.plot(x_points,fit_func(p_lsq_regularization[0], x_points),label='regularization')
plt.plot(x, y, 'bo', label='noise')
plt.legend()
```
