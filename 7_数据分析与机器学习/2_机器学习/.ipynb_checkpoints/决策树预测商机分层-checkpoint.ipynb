```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import graphviz
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取数据
path = r'C:\Users\caoyuanyuan\Desktop\项目文档\1-项目需求\数据\商机分层_1117\励销云特征分析_1117.xlsx'
df = pd.read_excel(path)

# 增加一列
df["共计访问周数"] = df["间断访问周数"] + df["最大连续访问周数"] - 1
df = pd.DataFrame(df,columns=['company_name','间断访问周数','最大连续访问周数','平均最大连续访问天数',
                            '平均最小连续访问天数','连续访问周数占比','最晚访问时间至今','平均访问小时数',
                            '平均每小时pv数','共计访问周数','是否成交客户'])
df.head()
```

```python
# 特征值
features = list(df.columns[[2,3,6,8,9]])
features
```

```python
X = np.array(df.iloc[:, [2,3,6,8,9]])
y = np.array(df.iloc[:, -1])
# 分割训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
model_tree = DecisionTreeClassifier(random_state=0)
model_tree.fit(X_train, y_train)
score = model_tree.score(X_test, y_test)
print(score)
```

```python
from sklearn import tree 
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["figure.figsize"]=(40, 40)
tree.plot_tree(model_tree,feature_names=features)
```

```python
## 预测数据
test_path = r'C:\Users\caoyuanyuan\Desktop\项目文档\1-项目需求\数据\商机分层_1117\励销云商机分析_1117.xlsx'
test_data = pd.read_excel(test_path, sheet_name='待预测数据')
test_column = np.array(test_data.iloc[:, [2,3,6,8,9]])
test_result = model_tree.predict(test_column)
test_data['test_result'] = test_result
test_data.head()
test_data.to_excel(r'C:\Users\caoyuanyuan\Desktop\项目文档\1-项目需求\数据\商机分层_1117\励销云商机结果预测_1117.xlsx')
```
