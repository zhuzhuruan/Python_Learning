# 数据仓库

# 1 概念

- <font color=#FF2121>数据仓库</font>（英语：Data Warehourse,简称数仓、DW），是用于存储、分析、报告的数据系统。
- 数据仓库的目的是构建<font color=#FF2121>面向分析</font>的集成化数据环境，分析结果为企业提供决策支持（Decision Support）。
- 数据仓库<font color=#FF2121>本身并不“生产”任何数据</font>，其数据来源于不同的外部系统。
- 同时数据仓库自身<font color=#FF2121>也不需要“消费”任何的数据</font>，其结果开放给各个外部应用使用。
- 这也是为什么叫“仓库”，而不叫“工厂”的原因。

# 2 数据仓库为何而来

先下结论：为了分析数据而来，分析结果给企业决策提供支撑。
企业中，信息总是用作两个目的：
<br>
（1）操作型记录的保存
<br>
（2）分析型决策的制定

**案例：以中国人寿保险公司发展为例，阐述数据仓库为何而来？**
<br>
**（1）操作型记录的保存**
- 中国人寿保险（集团）公司下辖多条业务线，包括：人寿险、财险、车险、养老险等。各业务线的业务正常运营需要记录维护包括客户、保单、收付费、核保、理赔等信息。
- <font color=#FF2121>联机事务处理系统（OLTP）</font>正好可以满足上述业务需求开展，其主要任务是执行联机事务处理。其基本特征是前台接收的用户数据可以立即传送到后台进行处理，并在很短的时间内给出处理结果。
- <font color=#FF2121>关系型数据库（RDBMS）是OLTP典型应用</font>，比如：Oracle、Mysql、SQL Sever等。
<br>

**（2）分析型决策的制定**
<br>
- 为了能够正确认识出现的问题，制定相关的解决措施，瞎拍桌子是肯定不行的。
- 最稳妥的办法就是：<font color=#FF2121>基于业务数据开展数据分析，基于分析的结果给决策提供支撑</font>。也就是所谓的数据驱动决策的制定。
<br>

**OLTP环境开展分析可行吗？**
<br>

**<font color=#FF2121>可以，但是没必要</font>**
- OLTP系统的核心是面向业务、支持业务、支持事务。所有的业务操作可以分为读、写两种操作，一般来说<font color=#FF2121>读的压力明显大于写的压力</font>。如果在OLTP环境直接开展各种分析，有以下问题需要考虑：
    - 数据分析也是对数据进行读取操作，会让读取压力倍增；
    - OLTP仅存储数周或数月的数据；
    - 数据分散在不同系统不同表中，字段类型属性不统一；
- 当分析所涉及数据规模较小的时候，在业务低峰期时可以在OLTP系统上开展直接分析。但是<font color=#FF2121>为了更好的进行各种规模的数据分析，同时也不影响OLTP系统运行，此时需要构建一个集成统一的数据分析平台</font>。
- 该平台的目的很简单：**面向分析，支持分析**，并且和OLTP系统解耦合。
- 基于这种需求，数据仓库的雏形开始在企业中出现了。
<br>

**数据仓库的构建**
- 如数据仓库定义所说，数仓是一个用于存储、分析、报告的数据系统，目的是构建面向分析的集成化数据环境。我们把这种<font color=#FF2121>面向分析、支持分析的系统称之为OLAP（联机分析处理）系统</font>。数据仓库是OLAP一种。
- 中国人寿保险公司就可以基于分析决策需求，构建数仓平台。
<br>

# 3 主要特征

- 面向主题（Subject-Oriented）
    - 主题是一个抽象的概念，是较高层次上数据综合、归类并进行分析利用的抽象。
- 集成性（Integrated）
    - 主题相关的数据通常会分布在多个操作型系统中，彼此分散、独立、异构。需要集成到数仓主题下。
- 非易失性（Non-Volatile）
    - 也叫非易变性。数据仓库是数据分析的平台，而不是创造数据的平台。
- 时变性（Time-Variant）
    - 数据仓库的数据需要随着时间更新，以适应决策的需要。

# 4 OLTP、OLAP

## 4.1 联机事务处理OLTP（On-Line Transaction Processing）

- 操作型处理，主要目标是做数据处理，它是针对具体业务在数据库联机的日常操作，通常对少数记录进行查询、修改
- 用户较为关心的响应时间、数据的安全性、完整性和并发支持的用户数等问题。
- <font color=#FF2121>传统的关系型数据库系统（RDBMS）作为数据管理的主要手段</font>，主要用于操作型处理。

## 4.2 联机分析处理OLAP（On-Line Analytical Processing）

- 分析型处理，主要目标是做数据分析。
- 一般针对某些主题的历史数据进行复杂的多维分析，支持管理决策。
- <font color=#FF2121>数据仓库是OLAP系统的一个典型示例</font>，主要用于数据分析。

## 4.3 对比


| |  **OLTP**   | **OLAP**  | 
|  :----:  |  :----:  |  :----:  |
| **数据源** | 仅包含当前运行日常业务数据  | 整合来自多个来源的数据，包括OLTP和外部来源  |
| **目的** | 面向应用、面向业务、支撑业务  | 面向主题、面向分析、支撑分析决策  |
| **焦点** | 当下  | 主要面向过去、面向历史（实时数仓除外）  |
| **任务** | 读写操作  | 大量读而很少写操作  |
| **响应时间** | 毫秒  | 秒、分钟、小时或者天（取决于数据量和查询复杂性） |
| **数据量** | 小数据，MB、GB  | 大数据，TB、PB  |

## 5 数据仓库、数据库

- 数据库与数据仓库的区别实际讲的是OLTP和OLAP的区别。
- OLTP系统的典型应用就是RDBMS，也就是我们俗称的数据库，当然这里要特别强调此数据库表示的是关系型数据库，Nosql数据库并不在讨论范围内。
- OLAP系统的典型应用就是DW，也就是我们俗称的数据仓库。
<br>
- 区别
    - 数据仓库不是大型的数据库，虽然数据仓库存储数据规模大。
    - 数据仓库的出现，并不是要取代数据库。
    - 数据库是面向事物的设计，数据仓库是面向主题设计的。
    - 数据库一般存储业务数据，数据仓库存储的一般是历史数据。
    - 数据库是为捕获数据而设计，数据仓库是为分析数据而设计。

# 6 数据仓库、数据集市

- 数据仓库（Data Warehouse）是面向整个集团组织的数据，数据集市（Data Mart）是面向单个部门使用的。
- 可以认为数据集市是数据仓库的子集，也有人<font color=#FF2121>把数据集市叫做小型数据仓库</font>。数据集市通常只涉及一个主题领域，例如市场营销或销售。因为它们较小且更具体，所以它们通常更易于管理和维护，并有更灵活的结构。
- 各种操作型数据库和包括文件在内的等其他数据作为数据源，经过ETL填充到数据仓库中；数据仓库中有不同的主题数据，数据集市则根据部门特点面向指定主题，比如Purchasing（采购）、Sales（销售）、Inventory（库存）。
- 用户可以基于主题数据开展各种应用：数据分析、数据报表、数据挖掘。

# 7 数据仓库分层架构

## 7.1 分层思想和标准

- 数据仓库的特点是本身不生产数据，也不最终消费数据。按照数据流入流出数仓的过程进行分层就显得水到渠成。
- 每个企业根据自己的业务需求可以分成不同的层次。但是最基础的分层思想，理论上分为三个层：操作型数据层（ODS）、数据仓库层（DW）和数据应用层（DA）。
- 企业在实际运用中可以基于这个基础分层之上添加新的层次，来满足不同的也无需求。

## 7.2 阿里巴巴数仓三层架构

- 阿里数仓是非常经典的3层架构，从下往上依次是：ODS、DW、DA。
- 通过元数据管理和数据质量监控来把整个数仓中数据的流转过程、血缘依赖关系和生命周期。
<br>

| 源数据   | ETL | 数据仓库  | 数据应用  | 元数据管理、数据质量监控
|  :----:  |  :----:  | :----:  | :----:  |:----:  |
| 点击流数据 | | 明细数据（DWD）  | KPI |  |
| 数据库数据 | | 低粒度汇总加工数据（DWB）  | 账单应用 |  | 
| 文档数据 | | 高粒度汇总加工数据（DWS）  | 日志产品应用 |  |
| 其他数据 | | 数据集市、宽表（DM）  | 报表展示 |
|  |  |数据应用（ST） | 自定义查询、数据分析、数据挖掘 |  |

- ODS层（Operation Data Store）
    - <font color=#FF2121>**操作型数据层**</font>，也称之为源数据层、数据引入层、数据暂存层、临时缓存层。
    - <font color=#E36C07>此层存放未经过处理的原始数据至数据仓库系统，结构上和源数据保持一致，是数据仓库的数据准备区</font>。
    - 主要完成基础数据引入到数仓的职责，和数据源系统进行解耦合，同时记录基础数据的历史变化。
- DW层（Data Warehouse）
    - <font color=#FF2121>**数局仓库层**</font>，由ODS层数据加工而成。主要完成<font color=#FF2121>数据加工与整合</font>，建立一致性的维度，构建可复用的面向分析和统计的明细事实表，以及汇总公共粒度的指标。内容具体划分如下：
        - 公共维度层（DIM）：基于维度建模理念思想，建立整个企业一致性维度。
        - 公共汇总粒度事实层（DWS、DWB）：以分析的主题对象作为建模驱动，基于上层的应用和产品的指标需求，构建公共粒度的汇总指标事实表，以宽表化手段物理化模型。
        - 明细粒度事实层（DWD）：将明细事实表的某些重要维度属性字段做适当冗余、即宽表化处理。
- DA层（或ADS层）
    - <font color=#FF2121>**数据应用层**</font>，面向最终用户，面向业务定制<font color=#FF2121>提供给产品和数据分析使用的数据</font>。
    - 包括前端报表、分析图表、KPI、仪表盘、OLAP专题、数据挖掘等分析。

## 7.2 好处

分层的主要原因是在管理数据的时候，能对数据有一个更加清晰的掌控，详细来讲，主要有下面几个原因：
<br>
- <font color=#FF2121>清晰数据结构</font>
    - 每一个数据分册都有它的作用域，在使用表的时候能更方便底定位和理解
- <font color=#FF2121>数据血缘追踪</font>
    - 简单来说，我们最终给业务呈现的是一个能直接使用业务表，但是它的来源有很多，如果有一张表出现了问题，我们希望能够快速准确地定位到问题，并清楚它的危害范围。
- <font color=#FF2121>减少重复开发</font>
    - 规范数据分层，开发一些通用的中间层数据，能够减少极大的重复计算。
- <font color=#FF2121>把复杂问题简单化</font>
    - 将一个复杂的任务分解成多个步骤来完成，每一层只处理单一的步骤，比较简单和容易理解。而且便于维护数据的准确性，当数据出现问题之后，可以不用修复所有的数据，只需要从有问题的步骤开始修复。
- <font color=#FF2121>屏蔽原始数据的异常</font>
    - 屏蔽业务的影响，不必改一次业务就炫耀重新接入数据。

#  8 ETL、ELT

- 背景
    - 数据仓库<font color=#E36C07>从各数据源获取数据及在数据仓库内的数据转换和流动</font>都可以认为是ETL（<font color=#E36C07>抽取Extra、转化Transfer、装载Load</font>）的过程。
    - 但是在实际操作中将数据加载到仓库却产生了两种不同的做法：ETL和ELT。
- Extract、Transform、Load ETL
    - 首先从数据源池中提取数据，这些数据源通常是事务性数据库。数据保存在临时暂存数据库中（ODS）。然后执行转换操作，将数据结构化并转换为适合目标数据仓库系统的形式。然后将结构化数据加载到仓库中，以备分析。
- Extract、Load、Transform ELT
    - 使用ELT，数据在从源数据池中提取后立即加载。没有专门的临时数据库（ODS），这意味着数据会立即加载到单一的集中存储库中。数据在数据仓库系统中进行转换，以便与商业智能工具（BI工具）一起使用。<font color=#FF2121>**大数据时代的数仓这个特点很明显**</font>
