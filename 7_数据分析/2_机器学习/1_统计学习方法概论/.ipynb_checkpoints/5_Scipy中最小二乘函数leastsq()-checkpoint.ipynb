# 一、含义

最小二乘法可以说是最经典的数值优化技术了， 通过最小化误差的平方来寻找最符合数据的曲线。在optimize模块中，使用leastsq（）函数可以很快速地使用最小二乘法对数据进行拟合。

作用：最小化一组方程的平方和。

函数调用格式：

scipy.optimize.leastsq(func, 
                       x0, 
                       args=(), 
                       Dfun=None, 
                       full_output=0, 
                       col_deriv=0, 
                       ftol=1.49012e-08, 
                       xtol=1.49012e-08, 
                       gtol=0.0, 
                       maxfev=0, 
                       epsfcn=None, 
                       factor=100, 
                       diag=None)

参数设置：

- func：误差函数
- x0：初始化的参数
- args：其他的额外参数

leastsq，它可以省去最小二乘法中间具体的求解步骤，只需要输入一系列样本点，给出待求函数的基本形状（例如二元二次函数就是一种形状$f(x,y)=w_{0}x^2 + w_{1}y^2 + w_{2}xy + w_{3}x + w_{4}y + w_{5}$，在形状给定后，我们只需要求解相应的系数w0~w5），即可得到相应的参数。

# 二、例子

## 函数形为y=kx+b

```python
import numpy as np
from scipy.optimize import leastsq
from matplotlib import pyplot as plt
```

```python
# 采样点
x=np.array([8.19,2.72,6.39,8.71,4.7,2.66,3.78])
y=np.array([7.01,2.78,6.47,6.71,4.1,4.23,4.05])

# 需要拟合的函数以及误差
def fit_func(p, x):
    k,b = p
    return k*x+b

def error(p,x,y):
    return fit_func(p,x)-y    #x、y都是列表，故返回值也是个列表
```

```python
#试验最小二乘法函数leastsq得调用几次error函数才能找到使得均方误差之和最小的k、b
p0 = [100, 2]
Para=leastsq(error,p0,args=(x,y)) #把error函数中除了p以外的参数打包到args中
k, b = Para[0]
# 绘图查看拟合效果
plt.figure(figsize=(8,6))
plt.scatter(x,y,color="red",label="Sample Point",linewidth=3) #画样本点
x1=np.linspace(0,10,1000)
y1=k*x1+b     # 通过最小二乘法求出的参数
plt.plot(x1,y1,color="orange",label="Fitting Line",linewidth=2) #画拟合直线
plt.legend()
plt.show()
```

**注意：**

- p0里放的是k、b的初始值，这个值可以随意指定。往后随着迭代次数增加，k、b将会不断变化，使得error函数的值越来越小。
- func函数里指出了待拟合函数的函数形状。
- error函数为误差函数，目标就是不断调整k和b使得error不断减小。这里的error函数和神经网络中常说的cost函数实际上是一回事，只不过这里更简单些而已。
- 必须注意一点，传入leastsq函数的参数可以有多个，但必须把参数的初始值p0和其它参数分开放。其它参数应打包到args中。

## 函数为多项式

```python
# 样本点
x=np.array([0,1,2,3,-1,-2,-3])
y=np.array([-1.21,1.9,3.2,10.3,2.2,3.71,8.7])

# 函数和误差
def fit_func(p, x):
    f = np.poly1d(p)   # 生成多项式函数，np.poly1d([1,2,3]) = 1x^2+2x^1+3x^0
    return f(x)

def error(p,x,y):
    return fit_func(p,x)-y

# 初始化参数
p0 = [5, 2, 10]
para = leastsq(error, p0, args=(x,y))
w1,w2,w3 = para[0]

# 绘图查看拟合效果
plt.figure(figsize=(8,6))
plt.scatter(x,y,color="red",label="Sample Point",linewidth=3) #画样本点
x1=np.linspace(0,10,1000)
y1=w1*x1**2+w2*x1+w3     # 通过最小二乘法求出的参数
plt.plot(x1,y1,color="orange",label="Fitting Line",linewidth=2) #画拟合直线
plt.legend()
plt.show()
```
