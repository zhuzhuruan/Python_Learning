# 数据指标体系

# 1 什么是数据指标

- 对当前业务有参考价值的统计数据
    - 当前业务 有价值：不是所有数据都叫指标，与当前业务无关的，也意义不大
    - 可统计：数据不是凭空产生的、不能脱离现实
- 常用的数据指标
    - 用户数据
    - 行为数据
    - 业务数据

# 1.1 用户数据
- 存量：DAU(日活) / MAU(月活)
- 增量：新增用户
- 健康程度：留存率
- 从哪儿来：渠道来源

## 1.1.1 理解Daily/Monthly

- Daily—自然日
    - Tips：跨时区（如全球服务），则关心最近24h

- Monthly—当月至少活跃一次的用户总数
    - Tips：MAU ≠ 当月DAU之和
    - 务必去重，才有观察的意义
<br>
<br>
- ***案例：一款只有10名用户的产品***
<br>
DAY1：DAU=10&ensp;&ensp;DAY2：DAU=8&ensp;&ensp;DAY3：DAU=5&ensp;&ensp;DAY1：DAU=10&ensp;&ensp;&ensp;...&ensp;&ensp;&ensp;DAY30：DAU=7
<br>
MAU  = 10
<br>
<font color=#E36C07>**Tips：MAU < 最大日活 + 新增用户 ≤ 总用户量**<font>

## 1.1.2 理解Active

- **数据统计系统的定义**
    - 预制报表的统计系统（友盟、百度统计、GA、......）都是基于***事件上报*** 进行统计
        - √用户申请打开网页：页面加载成功，上报Page View事件
        - ×后台回报消息：收到PUSH，上报事件
        - √用户点击按钮：按钮被点击，上报onclick事件
<br>
    - <font color=#E36C07>今天上报过事件 （→ **用户**进行了**主动**操作）→ 这个用户是活跃的<font>
- **业务上的定义**
    - 存在维护成本：需不断维护***日活事件列表***
        - 访问首页
        - 访问商品详情页
        - 访问抽奖页
        - .......
    - 存在沟通成本：团队内外对 ***活跃*** 的认识需统一
- **总结**
    - 方法一：数据统计系统的定义
        - 基于事件上报：有事件上报 → 该用户活跃
    <br>
        Tips：上报一事可能有“坑”（假定了事件上报一定来自用户主动操作）
        - 谨防surprise！（活跃暴增，其他数值并无显著增加）
    - 方法二：业务上的定义
        - 基于***关键事件上报***：用户执行了***关键事件*** → 该用户活跃
        - 存在维护成本（需不断维护日活事件列表）
        - 存在沟通成本（团队内外对活跃的认识需统一）

## 1.1.3 理解USER
- 认人
    - 给每位注册用户一个唯一的专属ID
    - 只适合强注册/登录环境，未登录的用户会被漏掉
    - <font color=#E36C07>用户数 = 访问过服务的ID数<font>
- 认设备
    - 在网页cookie中埋下一段长随机字符串，作为设备的唯一标识符
    - 无法对应设备背后的用户
    - <font color=#E36C07>用户数 = 访问过服务的设备数<font>
- 认人 or 认设备？
    - 是否有账号体系？
        - NO → 认设备
    - 业务场景是否强依赖登录？
        - YES → 认人 + 认设备
    - 不登录的用户对业务是否有价值？
        - NO → 认人 + 认设备 
        - YES → 认设备

## 1.1.4 如何定义新增
***案例：如何理解新增用户***
<br>
渠道商：“只要点了你们产品的下载按钮，就算一次新增了。”<br>
运营：“总得下载成功了才算吧！而且一个用户一天最多算一次。”<br>
产品：“我们按启动量算吧，没打开APP也算，数据质量太差了。”<br>
RD：“注册了才叫新增用户，否则后台压根没这条数据，你们怎么YY也没用。”<br>

- 问题一：选择合适的节点，定义***增***
<br>

| |  点击渠道链接   | 下载  | 安装/启动  | 激活  |
|  :----:  |  :----:  | :----:  |:----:  | :----:  |
| **优势** | 统计简单  | 真实反映了用户的实际意愿 | 离激活最近，便于统计 | 最真实的数据
| **劣势** | 离激活环节最远<br>转化率太差  | 数据源可信度存疑<br>无法避免刷量 | 渠道不一定配合<br>任然无法避免刷量 | 渠道费用激增<br>统计复杂
| **适用场景** | 量级不大/免费渠道<br>不需要做精细结算  | 渠道依赖应用商店<br>且没有更好的渠道 | 自己较强势，可给<br>渠道制定统计规则 | 对用户质量要求很<br>高且产品ARPU高|

- 问题二：用适当的方法，判别***新***
    - 基于设备：ios、android、web各有门道
    - 基于账号关联：与后台已有账号比对匹配

## 1.1.5 用户留存

***案例：一款只有10名用户的产品，7日留存的算法***
<br>
DAY1：DAU=10&ensp;&ensp;DAY2：DAU=8&ensp;&ensp;DAY3：DAU=5&ensp;&ensp;DAY1：DAU=10&ensp;&ensp;&ensp;...&ensp;&ensp;&ensp;DAY30：DAU=7
<br>
MAU  = 10
<br>
- 算法一：7日日留存（只关心到特定日的留存情况，避免了其他日数据的干扰）
    $$\frac{第七天}{第一天}\times100\%$$
    - 次日留存 $$\frac{第二天}{第一天}\times100\%$$
    - 三日留存 $$\frac{第三天}{第一天}\times100\%$$
<br>
- 算法二：7日内留存（引入了其他日数据，适用于有固定使用周期，且周期较长的业务）
    $$\frac{第二天{\sim}第七天去重后}{第一天}\times100\%$$
<br>
- 算法三：7日日留存（新增当日为第0日，下一日为第1日，使第7日与新增当日对齐，试图抵消某些星期级别的周期性差异）
    $$\frac{第七天}{第0天}\times100\%$$
    - 首日留存 $$\frac{第一天}{第0天}\times100\%$$
    - 次日留存 $$\frac{第二天}{第0天}\times100\%$$

**为什么要看留存？**
- 了解某一个渠道的质量——日留存
    - 以天为单位，衡量这个渠道来的用户当下&接下来的表现
    - 以\[×日日留存\]作为比较标准，可以避免其他日数据的干扰
- 观察整个大盘——周留存/月留存
    - 以周/月为单位，衡量产品的健康情况，观察用户在平台上的黏性
    - <font color=#E36C07>**务必去重!**<font>
    
- 公式
    - 7日日留存 $$\frac{第7天日活跃用户数}{第1天日活跃用户数}\times100\%$$
    - 30日日留存 $$\frac{第30天日活跃用户数}{第1天日活跃用户数}\times100\%$$
    - 次周周留存（**务必去重**） $$\frac{下周周活跃用户数}{当周周活跃用户数}\times100\%$$
    - 次月月留存（**务必去重**） $$\frac{下月月活跃用户数}{当月月活跃用户数}\times100\%$$

# 1.2 行为数据
- 次数/频率：PV、UV、访问深度
- 路径走通程度：转化率
- 做了多久：时长
- 质量：弹出量

**案例：某个网站的一天**
- 5个用户来到某官网，总共访问了13次首页
- 13次首页访问中，有8次都会点开课程详情
- 最后有2个用户注册了会员
- 这2人中后续有1个用户购买了课程

**问题：**
- A：“来了多少用户？访问情况如何？” --------- “来了5个，访问13次，人均2.6次”
- B：“首页转化效果如何？” --------- “有8/13=62%的人进入课程详情页”
- C：“有多少注册了的？” --------- “今儿来的5个人里有2个注册的，占40%”
- D：“注册了的又有多少当时付费的？” --------- “这2人只有1个付费了的，占50%”


## 1.2.1 PV/UV

- Page Views：~~页面浏览量~~（次数）
- Unique Visitors：~~独立访问数~~（人数）

**案例：浏览详情页 $\Rightarrow$ 发表评论**  <br>
- 详情页的评论转化率 = 1 / 详情页PV × 评论发表PV  <br>
    - 表示 ***详情页*** 引发讨论的能力
- 用户的评论转化率 = 1 / 详情页PV × 评论发表PV
    - 表示 ***用户*** 的评论倾向
- 人均行为次数 = 详情页PV/详情页UV；评论发表PV/评论发表UV
    - 表示人均页面查看次数
    - 表示人均评论次数

## 1.2.2 访问深度/访问时长/弹出率

- 访问深度：用户对产品的了解程度
    - 算法一：用户对某些关键行为的访问次数
    - 算法二：将网站内容/功能分成几个层级，以用户本次访问过最深的一级计算
- 访问时长
    - 通过统计特殊事件，支持业务需求
        - 统计视频被消费程度，评价内容质量（记录暂停/关闭页面后，播放器中视频进度条当前的位置）
- 弹出率（Bounce Rate）：用户只访问了一个页面就走的情况
    - <font color=#E36C07>**整个网站产生的所有会话的弹出率**<font>

# 1.3 业务数据
- 总量
    - GMV（Gross Merchandise Volume）
    - 访问时长
- 人均
    - ARPU（Average Revenue Per User）：总收入/总用户数
    - ARPPU（Average Revenue Per Paid User）：总收入/付费用户数
    - 人均访问时长
- 人数
    - 付费人数
    - 播放人数
- 健康程度
    - 付费率、付费频次
    - 观看率
- 被消费对象
    - SKU视角
    - 被消费内容视角

## 1.3.1 常用的业务数据指标

|  直接付费   | 适用场景  | 解决什么问题  | 非直接付费  |
|  :----:  | :----:  |:----:  | :----:  |
| GMV  | 总量 | 描述交易的金额总规模 | 目标完成数<br>(报名、点击、分享......)
| ARPU/ARPPU  | 人均 | 单个用户的贡献程度 | 人均访问时长
| 付费人数  | 人数 | 描述愿意为服务付费的人数总规模 | 完成人数|
| 付费率/付费频次  | 健康程度 | 描述总体上的用户付费意愿<br>评判一个服务的健康程度| 完成率|
| SKU视角  | 被消费对象 | 需要分析消费品本身的运营情况时 | 被消费内容视角|

# 2 选好数据指标的通用方法论

## 2.1 从业务的最终目的出发梳理业务模块

**案例：卖货**

$货品 \stackrel{通过}{\longrightarrow} 图文 \stackrel{卖}{\longrightarrow} 顾客$ <br>
$~~~~~~~~~~~~~~~~~~$$\uparrow$<br>
$~~~~~~~~~~~~~$图文工具<br>
$~~~~~~~~~~~~~~~~~~$$\uparrow$<br>
$~~~~~~~~~~~~~~~~$社区

我要通过***社区创作***的优美的***图文***来***卖货***<br>
我要通过***大量自媒体***高效创作***咨询***获取***广告收入***<br>
用户为了通过***便捷的学习系统***高效学习***课程***和***氛围良好的社群服务***提供的教学效果***支付学费***


- **常见的拆解角度**
    - 如何搞大/搞频繁（手段）
    - 往往有什么困难，我们通过什么特色方式解决的（工具）

## 2.2 判断业务模块所属类型

- 四类业务模块
    - 省时间
        - 工具模块
        - 交易模块
    - 杀时间
        - 内容浏览模块
        - 社区模块
    - 产品对用户的价值来自于产品本身
        - 工具模块
        - 内容浏览模块
    - 产品对用户的价值来自**连接**其他资源
        - 交易模块
        - 社区模块

## 2.3 根据业务模块所属类型选择数据指标

- 工具模块
    - 效率
- 内容浏览模块
    - 质&量
- 交易模块
    - 转化率
- 社区模块
    - 活跃

###  2.3.1 工具类模块关心的指标

| |  描述了什么   | 举个例子  | 做好了能怎么样  |
|  :----:  |  :----:  | :----:  |:----:  |
| **使用量** | 累积量，投入程度  | 拍照、笔记 | 用户黏性强 |
| **目标达成率** | 是否正常运转  | 支付、搜索 | 满意度高 | 
| **频次** | 能不能让用户养成习惯  | 闹钟 | 养成固定习惯 |


###  2.3.2 交易类模块关心的指标

| |  描述了什么   | 举个例子  | 做好了能怎么样  |
|  :----:  |  :----:  | :----:  |:----:  |
| **详情页转化率** | 核心场景转化效率 | 电商 | 更容易卖 |
| **金额** | 总的交易规模  | 电商、知识付费 | 卖更多 | 
| **客单价** | 单个用户价值  | 奢侈品海淘 | 卖更高价 |
| **复购率** | 收入的持久度  | 订阅式购物 | 卖更多次 |

###  2.3.3 内容浏览类模块关心的指标

| |  描述了什么   | 举个例子  | 做好了能怎么样  |
|  :----:  |  :----:  | :----:  |:----:  |
| **浏览数** | 累计量 | 头条类 | 有多少人阅读 |
| **浏览广度** | 覆盖内容库存情况  | 视频网站多个频道 | 库存利用效率更高 | 
| **浏览时长** | 占据用户多少时间  | 快手、抖音 | 减少竞品使用时间 |
| **内容互动** | 用户对内容的情感  | AB站 | 用户黏性 |

###  2.3.4 社区/社交类模块关心的指标

| |  描述了什么   | 举个例子  | 做好了能怎么样  |
|  :----:  |  :----:  | :----:  |:----:  |
| **发布量** | 用户创作内容的数量 | 贴吧、FB | 更多的话题源头 |
| **互动量** | 用户与用户互动的次数  | 微博 | 社区更具活力 | 
| **关系密度** | 用户与用户间的关系  | 微信 | 更有可能长期留存 |

# 3 实战：如何选择数据指标

**案例：闲鱼关注的数据指标**
- 我想通过给卖家提供易用的**快速发布工具**，给买家提供各种商品**信息浏览**方式，并通过“鱼塘”将双方聚集起来**分发信息**，从而承载C2C交易流量
    - 买家：各种商品**信息浏览**方式
        - 内容浏览：“分类”
        - 社区：“鱼塘”（按“地理位置”或“兴趣”，方便交流交易）
    - 卖家：**方便的商品发布工具**和**商品分发**方式
        - 工具：易用的快速发布工具
        - 社区：“鱼塘”用来分发
- 拆解指标
    - 内容浏览模块：分类
       - 商品浏览量
    - 社区模块：鱼塘
        - （商品）内容发布量/率
        - （商品）内容被查看量/率
        - （商品）内容被询问量/率
    - 工具模块：发布工具
        - 商品发布成功率
