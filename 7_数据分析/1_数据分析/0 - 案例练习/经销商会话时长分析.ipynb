1、回复状态：如果1个会话ID的最后1条发送方为"用户"，则返回"未回复"；如果1个会话ID的最后1条发送方为"商家"，则返回"已回复"

2、对话持续时间(s)：每个会话ID的最后1条消息发送时间减去第1条消息的发送时间

3、首次应答时间(s)：每个会话ID的第1次"商家"发送时间减去第1次"用户"发送时间

4、对话开始时间(s)：每个会话ID的第1次消息发送时间

5、平均回复时长(s)

```python
import numpy as np
import pandas as pd
```

```python
# 读取数据
df = pd.read_excel("./file/message.xlsx")
df.消息发送时间 = pd.to_datetime(df['消息发送时间'])
df.head(10)
```

```python
result = []
for (agecny, sess_id), data in df.groupby(["经销商", "会话ID"]):
    print(data)
#     if agecny == '佩戴' and sess_id == 1:
    # 回复状态
    status = "已回复" if data.iloc[-1].发送方 == "商家" else "未回复"
    # 对话持续时间
    stay_time = (data.消息发送时间.max() - data.消息发送时间.min()).total_seconds()
    # 首次应答时间
    t = data.query('发送方 == "商家"')
    first_res_time = (t.iloc[0].消息发送时间 - data.query('发送方 == "用户"').iloc[0].消息发送时间).total_seconds() if t.shape[
                                                                                                          0] > 0 else None
    # 对话开始时间
    start_time = data.query('发送方 == "用户"').iloc[0].消息发送时间
    # 平均回复时长
    user_id_list = list(data.query('发送方 == "用户"').index)
    seller_id_list = list(data.query('发送方 == "商家"').index)
    print(user_id_list, seller_id_list)
    compare_start_end = []
    for i in range(0, len(user_id_list)):
        for j in range(0, len(seller_id_list)):     
            if i < len(user_id_list)-1:
                if (user_id_list[i] < seller_id_list[j]) & (seller_id_list[j] < user_id_list[i+1]):
                    compare_start_end.append((user_id_list[i], seller_id_list[j])) 
                    break
            else:
                if (user_id_list[i] < seller_id_list[j]):
                    compare_start_end.append((user_id_list[i], seller_id_list[j])) 
                    break
    print(compare_start_end)
    sum_time = 0
    for i in compare_start_end:
        print(data.iloc[530])
        _time = (data.iloc[i[1]].消息发送时间 - data.iloc[i[0]].消息发送时间).total_seconds()
        sum_time += _time
    average_time = sum_time/len(compare_start_end)
    result.append([agecny, sess_id, status, stay_time, first_res_time, start_time, average_time]) 

res = pd.DataFrame(result,columns=["经销商", "会话ID", "回复状态",
                                       "对话持续时间", "首次应答时间", "对话开始时间", "平均回复时长"])
res
```
